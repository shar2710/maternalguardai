```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
plt.rcParams["figure.figsize"]=(12, 6)

df=pd.read_csv('sample_data (1).csv')

print("sape:", df.shape)
df.head()
```

```python
missing=df.isnull().sum().sort_values(ascending=False)
print("missing values per column:\n", missing[missing > 0])
```

```python
sns.countplot(x='preeclampsia', data=df)
plt.title('class distribution: preeclampsia (1) vs healthy (0)')
plt.xlabel('preeclampsia')
plt.ylabel('count')
plt.show()

print(df['preeclampsia'].value_counts(normalize=True) * 100)
```

```python
df.describe().T
```

```python
selected_features=['age', 'bmi', 'blood_pressure_systolic', 'glucose_level', 'hemoglobin']

df[selected_features].hist(bins=30, figsize=(15, 8), color='green')
plt.suptitle("distribution of selected features")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 12))
corr=df.corr()
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("correlation matrix")
plt.show()
```

```python
for feature in ['bmi', 'blood_pressure_systolic', 'glucose_level', 'hemoglobin']:
    sns.boxplot(data=df, x='preeclampsia', y=feature)
    plt.title(f'{feature} v/s preeclampsia')
    plt.show()
```
