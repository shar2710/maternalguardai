```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
plt.rcParams["figure.figsize"]=(12, 6)

df=pd.read_csv('comprehensive_preeclampsia_dataset.csv')

print(df.shape)
df.columns
```

```python
sns.countplot(x='label', data=df, palette='inferno')
plt.title('preeclampsia (1) vs healthy (0)')
plt.xlabel('preeclampsia')
plt.ylabel('count')
plt.show()
```

```python
df.describe().T
```

```python
selected_features = [
    'systolic_bp', 'diastolic_bp', 'mean_arterial_pressure',
    'proteinuria_24h', 'protein_creatinine_ratio',
    'serum_uric_acid', 'serum_creatinine',
    'age', 'bmi', 'gestational_age',
    'chronic_hypertension', 'diabetes', 'nulliparous', 
    'multiple_pregnancy', 'headache', 
    'alt_sgpt', 'platelet_count',
    'visual_disturbance', 'epigastric_pain',
    'swelling_edema', 'hemoglobin'
]

df[selected_features].hist(bins=30, figsize=(18, 12), color='green', edgecolor='black')

plt.suptitle("Distribution of Selected Clinical Features", fontsize=16)
plt.subplots_adjust(top=0.93) 
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 12))
corr=df.corr()
sns.heatmap(corr, cmap='inferno', annot=False)
plt.title("correlation matrix")
plt.show()
```

```python
for feature in ['systolic_bp', 'diastolic_bp', 'mean_arterial_pressure',
    'proteinuria_24h', 'protein_creatinine_ratio',
    'serum_uric_acid', 'serum_creatinine',
    'age', 'bmi', 'gestational_age',
    'chronic_hypertension', 'diabetes', 'nulliparous', 
    'multiple_pregnancy', 'headache', 
    'alt_sgpt', 'platelet_count',
    'visual_disturbance', 'epigastric_pain',
    'swelling_edema', 'hemoglobin']:
    sns.boxplot(data=df, x='label', y=feature)
    plt.title(f'{feature} v/s preeclampsia')
    plt.show()
```
